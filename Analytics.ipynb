```python
# Задание 3:График с датами покупок в разрезе по месяцам (сколько покупок было сделано в каждом месяце)
```

```python
data['Дата'] = pd.to_datetime(data['Дата'], yearfirst = True)
```

```python
data['month'] = [d.month for d in data['Дата']]
```

```python
data['month'].replace({1: "Январь", 2: "Февраль", 3: "Март", 4: "Апрель", 5: "Май", 6: "Июнь", 7: "Июль"}, inplace=True)
```

```python
data['month'].value_counts(ascending=True).plot(kind='bar');
```
